## **Ταξίδι στο Ευρωπαϊκό Κοινοβούλιο: η πορεία της Ελλάδας στην Ευρωβουλή από το 1984 μέχρι σήμερα**

<div align="right">Καπετανάκη Δήμητρα, Μαρίνου Χρυσούλα</div>

*Ένας χρόνος από την επέτειο των 40 χρόνων των πρώτων ελληνικών εκλογών για το Ευρωπαϊκό Κοινοβούλιο, καθώς πλησιάζει και το τέλος της πενταετίας για την 11η εκλογική αναμέτρηση. Το 2021 η Ελλάδα έκλεισε τα 40 της χρόνια στη Ευρωπαική Ένωση. Ένας χρόνος μετά, και η υπόθεση του Qatargate κλονίζει την πίστη των πολιτών στους Ευρωπαικούς θεσμούς και κυρίως στο Ευρωπαϊκό Κοινοβούλιο.* 

Οι ευρωεκλογές αποτελούν ίσως τη σημαντικότερη έκφραση της δημοκρατίας στην Ευρώπη, καθώς το Ευρωπαϊκό Κοινοβούλιο αποτελεί το μοναδικό άμεσα εκλεγμένο σώμα της Ευρωπαϊκής Ένωσης. 
Ανατρέχουμε λοιπόν στο ιστορικό των ελληνικών κομμάτων στην Ευρωβουλή, στα γεγονότα που επηρέασαν την εκλογή τους και τις αποφάσεις που παιρνουν για τους ευρωπαίους “συμμάχους” τους.

### **Πώς λειτουργεί το Ευρωκοινοβούλιο;** 

Το Ευρωπαικό Κοινοβούλιο είναι ένα από τα νομοθετικά όργανα της Ευρωπαικής Ένωσης, τόσο με εποπτικές όσο και με δημοσιονομικές αρμοδιότητες. Μετά την αποχώρηση του Ηνωμένου Βασιλείου από την Ευρωπαϊκή Ένωση το 2018, το Ευρωπαϊκό Κοινοβούλιο αποτελείται πλέον από 705 βουλευτές εκ των 27 Κρατών-Μελών και εκλογές γίνονται ανα 5ετία με αναλογική εκπροσώπησή τους.

Η λειτουργία του στηρίζεται στη δημιουργία Ομάδων βουλευτών με παρόμοια ιδεολογική βάση και συμφωνία στη πολιτική γραμμή, χωρίς αυτό να σημαίνει σύμπλευση στις πολιτικές αποφάσεις. Οι ομάδες αυτές παραδοσιακά αποτελούνται από ευρωπαϊκά πολιτικά κόμματα, εθνικά κόμματα και βουλευτές, με κοινό καταστατικό. Οι Ευρωβουλευτές κάθονται ανα πολιτική ομάδα, οι οποίες απαρτίζονται από βουλευτές διαφορετικών χωρών της Ευρωπης, δίνοντας τους έτσι χαρακτήρα πέρα από τον εθνικό. Το σημερινό Κοινοβούλιο αποτελείται από 7 πολιτικές ομάδες και ανεξάρτητους βουλευτές.


### **Η Ελληνική οπτική: εθνικά κόμματα**

Όπως και στην εγχώρια πολιτική, οι τάσεις των ψηφοφόρων είναι αισθητές, καθώς τα κόμματα της Νέας Δημοκρατίας, του ΠΑΣΟΚ (Πανελληνιο Σοσιαλιστικό Κίνημα) και του ΚΚΕ (Κομμουνιστικό Κόμμα Ελλάδος) διατηρούν τη θέση τους διαχρονικά στις εκλογικές αναμετρήσεις με διακυμανσεις στις έδρες. 

Πιο συγκεκριμένα:

Το ΠΑΣΟΚ, έχει πάρει τις περισσότερες έδρες 2 φορες - το 1984 και το 1994, ενώ καταλαμβάνει κατά μέσο όρο το 1,2% των συνολικών εδρών του κοινοβουλίου

Το ΚΚΕ, αν και δεν έχει βγει πρώτο σε έδρες, κατείχε 5 φορες την 3η θέση, ενώ το 1989 δεν εκλέχθηκε καθόλου, και καταλαμβάνει κατά μέσο όρο το 0,7% των συνολικών εδρών του κοινοβουλίου.

Ο ΣΥΡΙΖΑ, με την εμφάνιση του το 2009 ως ενωτικό ψηφοδέλτιο και ύστερα το 2014 ως κόμμα, μπαίνει και αυτός στην Ευρωβουλή και διατηρεί τη θέση του μέχρι σήμερα, ενώ έγινε το κόμμα με τον μεγαλυτερο αριθμο εδρών για τις ευρωεκλογές του 2014, με 6 έδρες στην Κοινοβουλευτική Ομάδα του Ευρωπαϊκού Λαϊκού Κόμματος.

### **Παίρνοντας αποφάσεις**

Οι παλαιότερες και με μεγαλύτερα ποσοστά στο Ευρωπαϊκό Κοινοβούλιο ομάδες ήταν και είναι ο συνασπισμός των ομάδων της Κoιvoβoυλευτικής Ομάδας τoυ Ευρωπαϊκoύ Λαϊκoύ Κόμματoς (Χριστιαvoδημoκράτες) [EPP] και της Ομάδας της Προοδευτικής Συμμαχίας των Σοσιαλιστών και Δημοκρατών στο Ευρωπαϊκό Κοινοβούλιο [S&D].

Επίσης, ομάδες που διατηρούν μεγάλα νούμερα στη διάρκεια των περιόδων αυτών είναι η Συνομοσπονδιακή Ομάδα της Ευρωπαϊκής Ενωτικής Αριστεράς και η Ομάδα των Πρασίνων.

Τα ελληνικά κόμματα από νωρίς φαίνεται να διαλέγουν τις συμμαχίες τους με μικρές αποκλίσεις στα κόμματα τις αριστεράς. Η ένταξή τους παραμένει σταθερή σε συγκεκριμένες ομάδες που διαχρονικά αλλάζουν όνομα και πολιτικές, ανάλογα με τις περιόδους που έρχονται και παρέρχονται. Είναι σημαντικό, βέβαια, να επισημανθεί πως βουλευτές μπορούν να μην ακολουθήσουν την απόφαση του εθνικού τους κόμματoς και να μεταπηδήσουν σε διαφορετικές πολιτικές ομάδες, κάτι που στην περίπτωση της Ελλάδας δεν συμβαίνει συχνά.

Τα ελληνικά κόμματα και οι αντιστοιχίες τους σε ομάδες είναι μια γρήγορη ματιά στην σταθερότητα των αποφάσεων των κομμάτων. 

Η Νέα Δημοκρατία από την πρώτη της είσοδο στο Ευρωκοινοβούλιο σταθερά συνεργάζεται με την ομάδα του Ευρωπαϊκού Λαικού Κόμματος, στην οποία οι βουλευτές καταλαμβάνουν κατά μέσο όρο το 30% των μελών της ομάδας.

Το ΠΑΣΟΚ, που διαχρονικά εντάσσεται στην Προοδευτική Συμμαχία των Σοσιαλιστών και Δημοκρατών, αποτελεί το 28% των βουλευτών της ομάδας του. 


Στην περίπτωση της Συνομοσπονδιακής Ομάδας της Ευρωπαικής Αριστεράς, η οποία καταλαμβάνει κατά μέσο όρο ένα ποσοστό 5.6% στην πορεία όλων αυτών των χρόνων, παρουσιάζεται να απορροφά περισσότερα από ένα ελληνικά κόμματα, όπως 'Ο Συνασπισμός της Αριστεράς' μέχρι τη διάλυση του το 2003, το ΚΚΕ (για τις χρονιές 1984, 1994, 1999, 2004 και 2009) και στη συνέχεια το ΣΥΡΙΖΑ, στοιχείο που ανεβάζει την ελληνική παρουσία εντός της ομάδας διαχρονικά στο 0.7%. Αξιοπρόσεκτο είναι το γεγονός ότι, το ποσοστό των εθνικών κομμάτων μέσα στην ομάδα είναι μεγαλύτερο από το συνολικό ποσοστό της ομάδας που καταλαμβάνει η πολιτική ομάδα μέσα στο Ευρωκοινοβούλιο.


Λίγα είναι τα ελληνικά κόμματα που θα απομακρυνθούν από αυτό το μοτίβο και συνήθως είναι αυτά που θα απομακρυνθούν αργότερα και από την πολιτική σκηνή. Οι βουλευτές του ΚΚΕ και της Χρυσής Αυγής είναι τα μοναδικά κόμματα στην Ελλάδα που το 2014 και το 2019 θα μείνουν εκτός πολιτικής ομάδας, ως Μη Εγγεγραμμένοι. Ομάδες που σημειώνουν υψηλά ποσοστά στο κοινοβούλιο  και υπάρχουν από τα πρώτα του χρόνια, φαίνεται να μην προσελκύουν τους ίδιους αριθμούς βουλευτών, παρά τη δημοτικότητά τους στα υπόλοιπα κράτη μέλη. Χαρακτηριστικά παραδείγματα είναι η Ομάδα της Συμμαχίας Φιλελευθέρων και Δημοκρατών για την Ευρώπη  [ALDE], στην οποία δεν εμφανίζονται καθόλου Έλληνες βουλευτές και η Συμμαχία των Συντηρητικών και Μεταρρυθμιστών ή Ομάδα των Πράσινων [GREEN], όπως αναδείχνεται στο γράφημα.

### **Το μέλλον**

Λαμβάνοντας υπόψιν την πορεία που έχει ακολουθήσει μέχρι σήμερα η Ελληνική πολιτική σκηνή μέσα στο Ευρωκοινοβουλευτικό περιβάλλον, θα μπορούσε κανείς να πει ότι κάθε πρόβλεψη είναι αναμενόμενη. Παρόλα αυτά, οι τελευταίες πολιτικές εξελίξεις, όπως το σκάνδαλο Qatargate με επίκεντρο το Ευρωκοινοβούλιο, ο πόλεμος στην Ουκρανία, σε συνδυασμό με τις επερχόμενες εθνικές εκλογές τον Απρίλιο του 2023, καθιστούν το μέλλον αβέβαιο.
